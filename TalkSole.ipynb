```python
from urllib.request import urlopen, Request
from bs4 import BeautifulSoup
import os
import pandas as pd
y=2017
m=['01','02','03','04','05','06','07','08','09','10','11','12']
df = pd.DataFrame(columns=['標題','發文時間','鞋評內容'])
while y < 2020:

    for i in m:
        url = "https://talksole.blogspot.com/" + str(y) +  '/' + i
        r = Request(url)
        r.add_header("User-Agent", "Mozilla/5.0")
        response = urlopen(r)
        html = BeautifulSoup(response)
    
        shoes = html.find_all("div", class_="date-outer")
        if shoes == []:
            print(y,'年',i,'月','沒有內容')
            
        else:
            print(y,'年',i,'月',url)
            shoes_date_list = []
            shoes_title_list = []
            shoes_text_list = []
            for s in shoes:
                shoes_title = s.find_all("h3",class_="post-title entry-title")
                shoes_date  = s.find_all('h2',class_="date-header")
                shoes_text  = s.find_all('div',class_='post-body entry-content')
            
                for i in shoes_title:
                    aaa = i.find("a").text
                    shoes_title_list.append(aaa)
                for j in shoes_date:
                    bbb = j.find("span").text
                    shoes_date_list.append(bbb)
                for k in shoes_text:
                    shoes_text_list.append(k.text)

                a=0
                
            while True:
                try:
                    data = [shoes_title_list[a],shoes_date_list[a],shoes_text_list[a]]
                except:
                    break
                s = pd.Series(data,index=['標題','發文時間','鞋評內容'])
                print(s)
                df =df.append(s,ignore_index=True)
                df.to_csv('talksole.csv',index=False,encoding="utf-8-sig")        
                a +=1
            
    y +=1
```
