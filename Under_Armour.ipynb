```python
import requests
from bs4 import BeautifulSoup
import os
from urllib.request import urlretrieve,urlopen,Request,HTTPError
import warnings
import pandas as pd
import random
import time
import json
from time import sleep
warnings.filterwarnings('ignore')

df=pd.DataFrame(columns=['商品型號','商品網址','商品照片','商品標題','品牌','分類','男女款','價格(台幣)',
                         '商品介紹','主要顏色','次要顏色','主色比例','次色比例'])
url='https://www.underarmour.com/en-us/api/json-grid/g/39418?lsg=cb80c228-a0f3-4734-bd6f-c4493cf71d58&offset=0&limit=60&stackId=stack_header_main&stackIdx=0&t=&hiddenProducts=0&initialOffset=0'
headers = { 'Content-Type': 'application/x-www-form-urlencoded',
            'Origin': 'https://www.underarmour.com',
            'Referer': 'https://www.underarmour.com/en-us/mens/athletic-shoes/g/3967?lsg=37a29fc6-5883-41c9-a785-1553efb978e8',
            'Sec-Fetch-Mode': 'cors',
            'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/77.0.3865.90 Safari/537.36'
    }
response = requests.get(url, headers = headers).text
url_list=[]
img_list = []
idx_list = []
title_list = []
price_list=[]
type_list=[]
brand_list=[]
gender_list=[]
# context_list=[]
for i in range(21):
    try:
        shoes_url = json.loads(response[5:])["_embedded"]["results"][0]["products"][i]["_links"]["web:canonical"]["href"]
        response1 = requests.get(shoes_url, headers=headers).text
        html1=BeautifulSoup(response1)
        shoes=html1.find_all('li',class_='color-chip')
        for s in shoes:
            shoes_color = s['title'][-3:]
        
        
        
            
            shoes_title=html1.find('h1',class_="buypanel_producttitle").find('span').text
            shoes_type=html1.find('div',class_='buypanel_cattitle').text
            shoes_price=html1.find('span',class_='buypanel_productprice-value').text.replace('$',"").replace('.00','').replace('Sale price ','')
            shoes_idx=html1.find('span',class_='buypanel_productstyle-num').text + '-' + shoes_color
            shoes_pic='https://underarmour.scene7.com/is/image/Underarmour/' + shoes_idx + '_DEFAULT?scl=1&fmt=jpg&qlt=85&wid=1000&hei=1000&size=900,900&cache=on,off&bgc=ffffff&resMode=sharp2'
            
            idx_list.append(shoes_idx)
            url_list.append(shoes_url)
            title_list.append(shoes_title)
            price_list.append(shoes_price)
            type_list.append(shoes_type)
            img_list.append(shoes_pic)
            
#             print(shoes_context)
            time = random.randint(1, 3)
            print("等待時間", time)
            sleep(time)
    except AttributeError as e:
        print(e)
        continue
        
        
a=0
while True:
    try:
        data = [idx_list[a],url_list[a],img_list[a],title_list[a],type_list[a],price_list[a]]
    except:
        break
    s = pd.Series(data,index=['商品型號','商品網址','商品照片','商品標題','分類','價格(台幣)'])
    print(s)
    df =df.append(s,ignore_index=True)
    df.to_csv('E:/python/PyETL/UA/man/ua.csv',index=False,encoding="utf-8-sig")        
    a +=1
    
```
